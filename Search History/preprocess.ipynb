```python
def diffunder60mins(t1,t2):
    y2=t2.year
    M2=t2.month
    d2=t2.day
    h2=t2.hour
    m2=t2.minute
    y1=t1.year
    M1=t1.month
    d1=t1.day
    h1=t1.hour
    m1=t1.minute
    if(y1==y2 and M1==M2 and d1==d2):
        if(abs(m2-m1+60*(h2-h1))>=60):
            return False
        else:
            return True
    else:
        return False
```

```python
time_objects=[]
for i in range(len(time)):
    time_objects.append(datetime.strptime(time[i], '%b %d, %Y, %I:%M:%S %p %Z'))
```

```python
# time_objects[0].hour
```

```python
session_number=[0]
l=len(time_objects)
j=0
for i in range(1,l):
    if(diffunder60mins(time_objects[i-1],time_objects[i])==True):
        session_number.append(j)
    else:
        j=j+1
        session_number.append(j)
```
